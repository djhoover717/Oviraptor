# Define System class

```python
class System:
    def __init__(self, sysid=None):
        self.sysid = sysid
```

# Read in the data

```python
# here's the planetary system data from the archive
keys, vals = read_csv_file(data_file, k_index=0, v_index=1)

data = {}
for k in keys:
    data[k] = np.array(get_csv_data(k, keys, vals))
    

# grab a reference key
k0 = list(data.keys())[0]

print("Loaded data for {0} planets in {1} systems".format(len(data[k0]), len(np.unique(data["hostname"]))))
```

# Chen & Kipping 2017

```python
def mass_radius(mp):
    """
    mp : planet mass [Earth radii]
    """
    
    M = np.log10(mp)
    
    if mp < 2.04:
        C, S = (0.00346, 0.2790)
    elif mp < 132:
        C, S = (-0.0925, 0.589)
    elif mp < 26600:
        C, S = (1.25, -0.044)
    else:
        C, S = (-2.85, 0.881)
        
    return 10.**(C + M*S)
```

# RV multis

```python
rv_multis = []
mu = []

for i, star in enumerate(np.unique(data["hostname"])):
    use = data["hostname"] == star
    
    
    if (data["system_type"][use][0] == "Radial Velocity")*(np.sum(use) >= 3):
        
        s = System(sysid=star)
        
        # stellar parameters
        s.mstar = float(data["st_mass"][use][0])
        s.rstar = float(data["st_rad"][use][0])
        s.tstar = float(data["st_teff"][use][0])
        
        # planet parameters
        s.per = np.array(data["pl_orbper"][use], dtype="float")
        s.mp = np.array(data["pl_bmasse"][use], dtype="float")
        s.rp = np.empty_like(s.mp)
        
        for j, mp in enumerate(s.mp):
            s.rp[j] = mass_radius(mp)
        
        # system parameters
        s.mu = np.sum(s.mp)/s.mstar*(MEARTH/MSUN)
        
        # store info
        rv_multis.append(s)
        mu.append(s.mu)
        
        
order = np.argsort(mu)

rv_multis = [rv_multis[i] for i in order]
mu = [mu[i] for i in order]
```

```python
# color info
cmap = plt.cm.get_cmap("plasma")

cval = np.empty(len(rv_multis))
for i, s in enumerate(rv_multis):
    cval[i] = s.mstar
cval /= cval.max()


# now make the figure
plt.figure(figsize=(18,6))

for i, s in enumerate(rv_multis):
    for j, p in enumerate(s.per):
        plt.scatter(i, s.per[j], 20*s.rp[j], c=[cmap(cval[i])])
        plt.axvline(i, color="lightgrey", ls=":", zorder=0)
        plt.text(i, 0.05, s.sysid, rotation=90, va="bottom", ha="right", color="#555555", fontsize=12)

plt.xticks([])
plt.xlabel(r"$\leftarrow\ system-to-star\ mass\ ratio\ \rightarrow$", fontsize=24)
plt.yscale("log")
plt.yticks([1e0,1e1,1e2,1e3,1e4], fontsize=16)
plt.ylabel("period [days]", fontsize=24)
plt.savefig(MAINPATH + "Figures/rv_gallery.pdf", bbox_inches="tight")
plt.show()
```
