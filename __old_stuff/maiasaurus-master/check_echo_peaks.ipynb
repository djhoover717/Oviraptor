# Read in CKS data

```python
# READ IN CKS DATABASE
# documentation of csv file found at http://www.astro.caltech.edu/~howard/cks/column-definitions.txt

cks_file = MAINPATH + 'catalogues/clean_kepler_catalogue.csv'

cks_data = []
with open(cks_file) as infile:
    reader = csv.reader(infile)
    
    for row in reader:
        cks_data.append(row)
        
KEYS = cks_data[0]
VALUES = cks_data[1:]

def getkey(keyname,keys=KEYS,values=VALUES):
    '''
    keyname = (string) of column definition, see CKS documentation
    '''
    kid = keys.index(keyname)
    
    dummy = []
    for row in values:
        dummy.append(row[kid])
    
    return dummy
```

```python
# PULL RELEVANT QUANTITIES FROM CKS
cks = {}

cks['starname']    = getkey('id_starname')             # star id - 'K00082'
cks['koi']         = getkey('id_koicand')
cks['kic']         = getkey('id_kic')                  # Kepler Input Catalogue ID
cks['disposition'] = getkey('koi_disposition')         # status - 'CONFIRMED', 'CANDIDATE', 'FALSE POSITIVE', etc.
cks['falsepos']    = getkey('cks_fp')                  # CKS false positive designation - 'False' or 'True'
cks['snr']         = getkey('koi_model_snr')           # transit signal-to-noise
cks['npl']         = getkey('koi_count')               # number of planets in system

cks['period']      = getkey('koi_period')              # period (days)
cks['rprs']      = getkey('koi_ror')                   # planet-to-star radius ratio
cks['koi_duration']      = getkey('koi_duration')      # transit duration (hours)
cks['rp']      = getkey('iso_prad')                    # CKS+isochrone constrained planet radius (earth radius)
cks['sma']      = getkey('iso_sma')                    # CKS+isochrone constrained semi-major axis (AU)

cks['rstar']      = getkey('iso_srad')                 # CKS+isochrone constrained stellar radius (solar radius)
cks['mstar']      = getkey('iso_smass')                # CKS+isochrone constrained stellar mas (solar mass)
```

```python
# QUICKLY CLEAN DATA
KEY0 = 'starname'

def check_lengths(d, key0=KEY0):
    '''
    checks that all data arrays are the same length
    '''
    keys = d.keys()    
    L0   = len(d[key0])
        
    for k in keys:
        if len(d[k]) != L0: raise ValueError('inconsistent array length for array "%s"' %k)
            
    return None


def convert_to_arrays(d):
    keys = d.keys()
    dnew = {}
    
    for k in keys:
        dnew[k] = np.asarray(d[k])
        
    return dnew

    
    
def remove_missing_data(d, key0=KEY0):
    '''
    removes empty data cells
    '''
    keys = d.keys()
    L0   = len(d[key0])
    dnew = {}
    
    empty = np.zeros(L0)
    
    for k in keys:
        empty += d[k] == ''
    
    empty = empty > 0.0

    for k in keys:
        dnew[k] = d[k][~empty]

    check_lengths(dnew)
    
    return dnew


# CLEAN DATA
cks = convert_to_arrays(cks)
print('total number of queried objects =', len(cks[KEY0]))

cks = remove_missing_data(cks)
print('after removing missing data, number of objects =', len(cks[KEY0]))

check_lengths(cks)
```

```python
# CONVERT STRINGS TO FLOATS
def is_number(s):
    '''
    checks if input string s is a number
    '''
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def convert_floats(d):
    '''
    converts numerical arrays from strings to floats
    '''
    keys = d.keys()
    dnew = {}
    
    for k in keys:
        if is_number(d[k][0]):
            dnew[k] = np.asarray(d[k],dtype='float')
        else:
            dnew[k] = np.asarray(d[k])
    
    return dnew


# CONVERT STRINGS TO FLOATS (where applicable)
cks = convert_floats(cks)
```

# Calculate characteristic spacings

```python
class System:
    def __init__(self, sysid=None):
        
        self.sysid = sysid
```

```python
cks_uniquesys = np.unique(cks['starname'])
cks_systems   = []

for us in cks_uniquesys:
    # initialize System() object
    s = System()
    
    # pull stellar data
    s.sysid   = us
    s.Mstar   = cks['mstar'][cks['starname']==us][0]
    
    # pull planetary data
    s.masses  = cks['mp'][cks['starname']==us]
    s.periods = cks['period'][cks['starname']==us]

    
    # sort masses, periods, and durations by period
    order = np.argsort(s.periods)
    
    s.masses  = s.masses[order]
    s.periods = s.periods[order]

    
    # calcualte complexity metrics
    s.N  = len(s.masses)
    s.S  = archinfo.S(s.periods, s.masses, s.Mstar)
      
    cks_systems.append(s)
    
    

N_cks = np.zeros(len(cks_uniquesys))
S_cks = np.zeros(len(cks_uniquesys))

for i, s in enumerate(cks_systems):
    N_cks[i]  = s.N
    S_cks[i]  = s.S
```

```python
epos_all_uniquesys = np.unique(epos_all['star_id'])
epos_all_systems   = []

for us in epos_all_uniquesys:
    # initialize System() object
    s = System()
    
    # pull stellar data
    s.sysid   = us
    s.Mstar   = epos_all['mstar'][epos_all['star_id']==us][0]
    
    # pull planetary data
    s.masses  = epos_all['mass'][epos_all['star_id']==us]
    s.periods = epos_all['period'][epos_all['star_id']==us]

    
    # sort masses, periods, and durations by period
    order = np.argsort(s.periods)
    
    s.masses  = s.masses[order]
    s.periods = s.periods[order]

    
    # calcualte complexity metrics
    s.N  = len(s.masses)
    s.S  = archinfo.S(s.periods, s.masses, s.Mstar)
      
    epos_all_systems.append(s)
    
    

N_epos_all = np.zeros(len(epos_all_uniquesys))
S_epos_all = np.zeros(len(epos_all_uniquesys))

for i, s in enumerate(epos_all_systems):
    N_epos_all[i]  = s.N
    S_epos_all[i]  = s.S
    
    
bad = np.isnan(S_epos_all)

N_epos_all = N_epos_all[~bad]
S_epos_all = S_epos_all[~bad]
```
