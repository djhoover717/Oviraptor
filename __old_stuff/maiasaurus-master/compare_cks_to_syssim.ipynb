```python
# READ IN CKS DATABASE
# documentation of csv file found at http://www.astro.caltech.edu/~howard/cks/column-definitions.txt

cks_file = MAINPATH + 'catalogues/clean_kepler_catalogue.csv'

cks_data = []
with open(cks_file) as infile:
    reader = csv.reader(infile)
    
    for row in reader:
        cks_data.append(row)
        
KEYS = cks_data[0]
VALUES = cks_data[1:]

def getkey(keyname,keys=KEYS,values=VALUES):
    '''
    keyname = (string) of column definition, see CKS documentation
    '''
    kid = keys.index(keyname)
    
    dummy = []
    for row in values:
        dummy.append(row[kid])
    
    return dummy
```

```python
# PULL RELEVANT QUANTITIES FROM CKS
cks = {}

cks['starname']    = getkey('id_starname')             # star id - 'K00082'
cks['koi']         = getkey('id_koicand')
cks['kic']         = getkey('id_kic')                  # Kepler Input Catalogue ID
cks['disposition'] = getkey('koi_disposition')         # status - 'CONFIRMED', 'CANDIDATE', 'FALSE POSITIVE', etc.
cks['falsepos']    = getkey('cks_fp')                  # CKS false positive designation - 'False' or 'True'
cks['snr']         = getkey('koi_model_snr')           # transit signal-to-noise
cks['npl']         = getkey('koi_count')               # number of planets in system

cks['period']      = getkey('koi_period')              # period (days)
cks['period_err1'] = getkey('koi_period_err1')
cks['period_err2'] = getkey('koi_period_err2')

cks['rprs']      = getkey('koi_ror')                   # planet-to-star radius ratio
cks['rprs_err1'] = getkey('koi_ror_err1')
cks['rprs_err2'] = getkey('koi_ror_err2')

cks['impact']      = getkey('koi_impact')              # impact parameter
cks['impact_err1'] = getkey('koi_impact_err1')
cks['impact_err2'] = getkey('koi_impact_err2')

cks['koi_duration']      = getkey('koi_duration')      # transit duration (hours)
cks['koi_duration_err1'] = getkey('koi_duration_err1')
cks['koi_duration_err2'] = getkey('koi_duration_err2')

cks['rp']      = getkey('iso_prad')                    # CKS+isochrone constrained planet radius (earth radius)
cks['rp_err1'] = getkey('iso_prad_err1')
cks['rp_err2'] = getkey('iso_prad_err2')

cks['sma']      = getkey('iso_sma')                    # CKS+isochrone constrained semi-major axis (AU)
cks['sma_err1'] = getkey('iso_sma_err1')
cks['sma_err2'] = getkey('iso_sma_err2')

cks['rstar']      = getkey('iso_srad')                 # CKS+isochrone constrained stellar radius (solar radius)
cks['rstar_err1'] = getkey('iso_srad_err1')
cks['rstar_err2'] = getkey('iso_srad_err2')

cks['mstar']      = getkey('iso_smass')                # CKS+isochrone constrained stellar mas (solar mass)
cks['mstar_err1'] = getkey('iso_smass_err1')
cks['mstar_err2'] = getkey('iso_smass_err2')
```

```python
# QUICKLY CLEAN DATA
KEY0 = 'starname'

def check_lengths(d, key0=KEY0):
    '''
    checks that all data arrays are the same length
    '''
    keys = d.keys()    
    L0   = len(d[key0])
        
    for k in keys:
        if len(d[k]) != L0: raise ValueError('inconsistent array length for array "%s"' %k)
            
    return None


def convert_to_arrays(d):
    keys = d.keys()
    dnew = {}
    
    for k in keys:
        dnew[k] = np.asarray(d[k])
        
    return dnew

    
    
def remove_missing_data(d, key0=KEY0):
    '''
    removes empty data cells
    '''
    keys = d.keys()
    L0   = len(d[key0])
    dnew = {}
    
    empty = np.zeros(L0)
    
    for k in keys:
        empty += d[k] == ''
    
    empty = empty > 0.0

    for k in keys:
        dnew[k] = d[k][~empty]

    check_lengths(dnew)
    
    return dnew


# CLEAN DATA
cks = convert_to_arrays(cks)
print('total number of queried objects =', len(cks[KEY0]))

cks = remove_missing_data(cks)
print('after removing missing data, number of objects =', len(cks[KEY0]))

check_lengths(cks)
```

```python
# CONVERT STRINGS TO FLOATS
def is_number(s):
    '''
    checks if input string s is a number
    '''
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def convert_floats(d):
    '''
    converts numerical arrays from strings to floats
    '''
    keys = d.keys()
    dnew = {}
    
    for k in keys:
        if is_number(d[k][0]):
            dnew[k] = np.asarray(d[k],dtype='float')
        else:
            dnew[k] = np.asarray(d[k])
    
    return dnew


# CONVERT STRINGS TO FLOATS (where applicable)
cks = convert_floats(cks)
```

# Calculate compexity metrics for each system

```python
class System:
    def __init__(self, sysid=None):
        
        self.sysid = sysid
```

```python
uniquesys = np.unique(cks['starname'])
systems   = []

for us in uniquesys:
    # initialize System() object
    s = System()
    
    # pull stellar data
    s.sysid   = us
    s.Mstar   = cks['mstar'][cks['starname']==us][0]
    s.Rstar   = cks['rstar'][cks['starname']==us][0]
    
    # pull planetary data
    s.rprs    = cks['rprs'][cks['starname']==us]
    s.masses  = cks['mp'][cks['starname']==us]
    s.periods = cks['period'][cks['starname']==us]

    # pull transit durations and errors, convert hrs to days
    s.dur     = cks['koi_duration'][cks['starname']==s.sysid]/24
    s.dur_err = 0.5*((cks['koi_duration_err1'][cks['starname']==s.sysid]) -
                     (cks['koi_duration_err2'][cks['starname']==s.sysid]))/24
    
    # sort masses, periods, and durations by period
    order = np.argsort(s.periods)
    
    s.rprs    = s.rprs[order]
    s.masses  = s.masses[order]
    s.periods = s.periods[order]
    s.dur     = s.dur[order]
    s.dur_err = s.dur_err[order]
    
    
    # calculate complexity measures (N, mu, Q, M, S, C, f)
    s.N  = len(s.masses)
    s.mu = archinfo.mu(s.masses, s.Mstar)
    s.Q  = archinfo.Q(s.masses)
    s.M  = archinfo.M(s.periods, s.masses)
    s.S  = archinfo.S(s.periods, s.masses, s.Mstar)
    s.C  = archinfo.C(s.periods)
    s.f  = archinfo.f(s.periods, s.Mstar/s.Rstar**3, s.rprs, s.dur, s.dur_err)
      
    systems.append(s)
```

```python
Nsys = len(uniquesys)

N  = np.zeros(Nsys)
mu = np.zeros(Nsys)
Q  = np.zeros(Nsys)
M  = np.zeros(Nsys)
S  = np.zeros(Nsys)
C  = np.zeros(Nsys)
f  = np.zeros(Nsys)

for i, s in enumerate(systems):
    N[i]  = s.N
    mu[i] = s.mu
    Q[i]  = s.Q
    M[i]  = s.M
    S[i]  = s.S
    C[i]  = s.C
    f[i]  = s.f

logmu = np.log10(mu)
```

# Calculate complexity metrics for SysSim systems

```python
# determine number of planets per system
star_id = np.array(syssim['target_id'], dtype='float')

# count up the planets in each system
syssim_npl = np.zeros_like(star_id, dtype='int')

for i, sid in enumerate(star_id):
    syssim_npl[i] = np.sum(star_id == sid)
    
syssim['npl'] = syssim_npl

# remove single-planet systems
for k in syssim.keys():
    syssim[k] = syssim[k][syssim_npl > 1]
```

```python
# calculate system-level measures

syssim_uniquesys = np.unique(syssim['target_id'])
syssim_systems   = []

for us in syssim_uniquesys:   
    # initialize System() object
    s = System()
    
    # pull stellar data
    s.sysid   = us
    s.Mstar   = syssim['mstar'][syssim['target_id']==us][0]
    s.Rstar   = syssim['rstar'][syssim['target_id']==us][0]
    
    # pull planetary data
    s.rprs    = syssim['radius'][syssim['target_id']==us]/RSRE
    s.masses  = syssim['mass'][syssim['target_id']==us]
    s.periods = syssim['period'][syssim['target_id']==us]

    # pull transit durations and errors
    s.dur = syssim['tdur'][syssim['target_id']==s.sysid]
    
    # sort masses, periods, and durations by period
    order = np.argsort(s.periods)
    
    s.rprs    = s.rprs[order]
    s.masses  = s.masses[order]
    s.periods = s.periods[order]
    s.dur     = s.dur[order]
    s.dur_err = np.ones_like(s.dur)/48    # no tracked durations -- set at 30 min (= one LC)
    
    # calculate complexity measures (N, mu, Q, M, S, C, f)
    s.N  = len(s.masses)
    s.mu = archinfo.mu(s.masses, s.Mstar)
    s.Q  = archinfo.Q(s.masses)
    s.M  = archinfo.M(s.periods, s.masses)
    s.S  = archinfo.S(s.periods, s.masses, s.Mstar)
    s.C  = archinfo.C(s.periods)
    s.f  = archinfo.f(s.periods, s.Mstar/s.Rstar**3, s.rprs, s.dur, s.dur_err)
    
    syssim_systems.append(s)
```
